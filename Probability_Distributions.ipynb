## Part 1 (Binomial Distribution)

The binomial distribution is a probability distribution that models the number of successes in a fixed number of trials of a binary (yes/no) experiment. Each trial is independent, and each trial has the same probability of success.
### Relevant Question:
**"What is the probability of finding exactly 2 defective bulbs out of 15 tested bulbs, when the probability of each bulb being defective is 5%?"**

### Real-World Example:
**Quality Control in Manufacturing:**
Imagine a company that manufactures light bulbs. They want to ensure the quality of their products by testing a random sample of bulbs for defects before they are shipped to stores. The company knows from past data that, on average, **5% of the bulbs are defective**.

If you want to find the probability of exactly 2 defective bulbs (k = 2), you would use the binomial formula.
image.png

The collection of those probabilities from 0 to n(in this case 15) is what makes up the binomial distribution.


```python
import numpy as np
import matplotlib.pyplot as plt

def binomial_pmf(n, p, k=None):
    """Calculate the binomial probability mass function (PMF)."""
    if k is not None:
        # Directly calculate the PMF for exactly k defective bulbs
        return np.math.comb(n, k) * (p ** k) * ((1 - p) ** (n - k))
    else:
        # Calculate the full PMF for all possible outcomes (0 to n defective bulbs)
        return [np.math.comb(n, i) * (p ** i) * ((1 - p) ** (n - i)) for i in range(n + 1)]

def plot_binomial_distribution(n, probabilities, labels):
    """Plot binomial distributions for different probabilities."""
    x = np.arange(0, n + 1)

    # Create a plot for each probability
    plt.figure(figsize=(15, 5))
    for i, p in enumerate(probabilities):
        pmf = binomial_pmf(n, p)
        plt.subplot(1, len(probabilities), i + 1)
        plt.bar(x, pmf, color='b', alpha=0.7)
        plt.title(f'Binomial Distribution (n={n}, p={p})')
        plt.xlabel('Number of Defective Bulbs')
        plt.ylabel('Probability')
        plt.grid(True)

    plt.tight_layout()
    plt.show()

def calculate_exact_probability(n, p, k):
    """Calculate the probability of exactly k defective bulbs."""
    return binomial_pmf(n, p, k)

# Parameters
n = 15  # Number of trials (bulbs)
probabilities = [0.05, 0.5, 0.8]
labels = [f'p={p}' for p in probabilities]

# Plotting the binomial distributions for the different probabilities
plot_binomial_distribution(n, probabilities, labels)

# Calculating the probability of exactly 2 defective bulbs for p=0.05
prob_2_defective = binomial_pmf(n, probabilities[0], 2)
print(f"Probability of exactly 2 defective bulbs: {prob_2_defective:.4f}")
```

# Key Insights
1. The **Binomial distribution**, **Normal distribution**, **Poisson distribution**, and **Exponential distribution** are all important probability distributions, but they differ in terms of their properties, application, and the situations in which they are used. Lets compare them:

### **Key Differences:**

| **Characteristic**         | **Binomial**                              | **Normal**                          | **Poisson**                         | **Exponential**                    |
|----------------------------|-------------------------------------------|-------------------------------------|-------------------------------------|-------------------------------------|
| **Type of Distribution**    | Discrete                                  | Continuous                          | Discrete                            | Continuous                          |
| **Parameters**              | \( n \) (number of trials), \( p \) (probability of success) | \( \mu \) (mean), \( \sigma \) (standard deviation) | \( \lambda \) (rate or average number of events) | \( \lambda \) (rate, inverse of mean time) |
| **Use Case**                | Success/failure in fixed trials (coin flips, etc.) | Data clustering around a mean (heights, exam scores) | Number of events in a fixed interval (calls, accidents) | Time between events in a Poisson process |
| **Shape of Distribution**   | Symmetric (if \( p = 0.5 \)) or skewed (otherwise) | Bell-shaped (symmetric)             | Skewed (right-tail)                 | Skewed (right-tail)                 |
| **Formula**                 | Binomial PMF                             | Normal PDF                          | Poisson PMF                         | Exponential PDF                     |
| **Example**                 | Number of heads in 10 coin flips         | Heights of people                   | Number of cars passing a toll booth | Time between two customer arrivals  |

### Summary:
- **Binomial**: Discrete, with a fixed number of trials and a binary outcome.
- **Normal**: Continuous, bell-shaped distribution, used for data that tends to cluster around a mean.
- **Poisson**: Discrete, counts events that happen at a constant rate over time or space.
- **Exponential**: Continuous, models the time between events in a Poisson process.

2. **Would this question work with other distributions?**
- **Exponential Distribution**: If you're modeling a continuous outcome like the time until the first defective bulb appears, the Exponential distribution is indeed a better fit. This is because it describes the time between events in a Poisson process, where the events (defective bulbs) occur randomly and independently over time.

- **Poisson Distribution**: If you're modeling the number of defects in a very large number of bulbs where the probability of a defect per bulb is very small, the Poisson distribution works well. It approximates the Binomial distribution when the number of trials is large, and the probability of success (defect) is small. The Poisson distribution is often used for rare events over a fixed period or space.
