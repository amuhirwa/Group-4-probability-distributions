## Part 1 (Binomial Distribution)

image.png

The binomial distribution is a probability distribution that models the number of successes in a fixed number of trials of a binary (yes/no) experiment. Each trial is independent, and each trial has the same probability of success.
### Relevant Question:
**"What is the probability of finding exactly 2 defective bulbs out of 15 tested bulbs, when the probability of each bulb being defective is 5%?"**

### Real-World Example:
**Quality Control in Manufacturing:**
Imagine a company that manufactures light bulbs. They want to ensure the quality of their products by testing a random sample of bulbs for defects before they are shipped to stores. The company knows from past data that, on average, **5% of the bulbs are defective**.

If you want to find the probability of exactly 2 defective bulbs (k = 2), you would use the binomial formula.
image.png

The collection of those probabilities from 0 to n(in this case 15) is what makes up the binomial distribution.


```python
import numpy as np
import matplotlib.pyplot as plt

def binomial_pmf(n, p, k=None):
    """Calculate the binomial probability mass function (PMF)."""
    if k is not None:
        # Directly calculate the PMF for exactly k defective bulbs
        return np.math.comb(n, k) * (p ** k) * ((1 - p) ** (n - k))
    else:
        # Calculate the full PMF for all possible outcomes (0 to n defective bulbs)
        return [np.math.comb(n, i) * (p ** i) * ((1 - p) ** (n - i)) for i in range(n + 1)]

def plot_binomial_distribution(n, probabilities, labels):
    """Plot binomial distributions for different probabilities."""
    x = np.arange(0, n + 1)

    # Create a plot for each probability
    plt.figure(figsize=(15, 5))
    for i, p in enumerate(probabilities):
        pmf = binomial_pmf(n, p)
        plt.subplot(1, len(probabilities), i + 1)
        plt.bar(x, pmf, color='b', alpha=0.7)
        plt.title(f'Binomial Distribution (n={n}, p={p})')
        plt.xlabel('Number of Defective Bulbs')
        plt.ylabel('Probability')
        plt.grid(True)

    plt.tight_layout()
    plt.show()

def calculate_exact_probability(n, p, k):
    """Calculate the probability of exactly k defective bulbs."""
    return binomial_pmf(n, p, k)

# Parameters
n = 15  # Number of trials (bulbs)
probabilities = [0.05, 0.5, 0.8]
labels = [f'p={p}' for p in probabilities]

# Plotting the binomial distributions for the different probabilities
plot_binomial_distribution(n, probabilities, labels)

# Calculating the probability of exactly 2 defective bulbs for p=0.05
prob_2_defective = binomial_pmf(n, probabilities[0], 2)
print(f"Probability of exactly 2 defective bulbs: {prob_2_defective:.4f}")
```

# Key Insights
1. The **Binomial distribution**, **Normal distribution**, **Poisson distribution**, and **Exponential distribution** are all important probability distributions, but they differ in terms of their properties, application, and the situations in which they are used. Lets compare them:

### **Key Differences:**

| **Characteristic**         | **Binomial**                              | **Normal**                          | **Poisson**                         | **Exponential**                    |
|----------------------------|-------------------------------------------|-------------------------------------|-------------------------------------|-------------------------------------|
| **Type of Distribution**    | Discrete                                  | Continuous                          | Discrete                            | Continuous                          |
| **Parameters**              | \( n \) (number of trials), \( p \) (probability of success) | \( \mu \) (mean), \( \sigma \) (standard deviation) | \( \lambda \) (rate or average number of events) | \( \lambda \) (rate, inverse of mean time) |
| **Use Case**                | Success/failure in fixed trials (coin flips, etc.) | Data clustering around a mean (heights, exam scores) | Number of events in a fixed interval (calls, accidents) | Time between events in a Poisson process |
| **Shape of Distribution**   | Symmetric (if \( p = 0.5 \)) or skewed (otherwise) | Bell-shaped (symmetric)             | Skewed (right-tail)                 | Skewed (right-tail)                 |
| **Formula**                 | Binomial PMF                             | Normal PDF                          | Poisson PMF                         | Exponential PDF                     |
| **Example**                 | Number of heads in 10 coin flips         | Heights of people                   | Number of cars passing a toll booth | Time between two customer arrivals  |

### Summary:
- **Binomial**: Discrete, with a fixed number of trials and a binary outcome.
- **Normal**: Continuous, bell-shaped distribution, used for data that tends to cluster around a mean.
- **Poisson**: Discrete, counts events that happen at a constant rate over time or space.
- **Exponential**: Continuous, models the time between events in a Poisson process.

2. **Would this question work with other distributions?**
- **Exponential Distribution**: If you're modeling a continuous outcome like the time until the first defective bulb appears, the Exponential distribution is indeed a better fit. This is because it describes the time between events in a Poisson process, where the events (defective bulbs) occur randomly and independently over time.

- **Poisson Distribution**: If you're modeling the number of defects in a very large number of bulbs where the probability of a defect per bulb is very small, the Poisson distribution works well. It approximates the Binomial distribution when the number of trials is large, and the probability of success (defect) is small. The Poisson distribution is often used for rare events over a fixed period or space.


## Part 2: Bayesian Probability
image.png
### Real-World Problem: **Medical Diagnosis (Disease Detection)**

For this example, we'll use **Bayesian Probability** to model the probability of a person having a disease based on certain diagnostic test results. This is a classic use case of Bayesian inference where we can update our belief about a patient's condition based on prior information and new evidence (in this case, the test results).

### Problem Setup

- **Prior Probability (P(Disease))**: The initial belief about the probability of a person having the disease before we see the test results. For example, let's say the general population has a 1% chance of having the disease (P(Disease) = 0.01).
  
- **Likelihood (P(Test | Disease))**: The probability of getting a positive test result given that the person has the disease. This is the sensitivity of the test. For example, if the test correctly identifies the disease 95% of the time when the person has it, then P(Test | Disease) = 0.95.
  
- **False Positive Rate (P(Test | No Disease))**: The probability of getting a positive test result when the person does **not** have the disease. This is the test's false positive rate. For example, if the test incorrectly identifies the disease 5% of the time for healthy individuals, P(Test | No Disease) = 0.05.

- **Posterior Probability (P(Disease | Test))**: After observing the test result (e.g., a positive result), we want to update our belief about the probability of the person having the disease using **Bayes' theorem**.

### Bayes' Theorem:
image.png

### Step-by-Step Calculation:

1. **Prior (P(Disease))**: We start with the prior probability that a person has the disease, which is 1% (0.01).
   
2. **Likelihood (P(Test | Disease))**: The likelihood of getting a positive test result when the person has the disease, which is 95% (0.95).
   
3. **False Positive Rate (P(Test | No Disease))**: The probability of getting a positive test result when the person does not have the disease, which is 5% (0.05).
   
4. **Posterior (P(Disease | Test))**: After we observe the test result (positive), we calculate the posterior probability of the person having the disease using Bayes' theorem.

```python
# Define prior, likelihood, and false positive rate
P_disease = 0.01  # Prior probability of having the disease
P_no_disease = 1 - P_disease  # Complement of the prior

P_test_given_disease = 0.95  # Likelihood: Test positive given disease
P_test_given_no_disease = 0.05  # False positive rate: Test positive given no disease

# Bayes' Theorem to calculate the posterior probability
def bayes_theorem(P_test_given_disease, P_disease, P_test_given_no_disease, P_no_disease):
    # Calculate P(Test)
    P_test = P_test_given_disease * P_disease + P_test_given_no_disease * P_no_disease

    # Apply Bayes' theorem
    P_disease_given_test = (P_test_given_disease * P_disease) / P_test

    return P_disease_given_test

# Calculate the posterior probability
posterior = bayes_theorem(P_test_given_disease, P_disease, P_test_given_no_disease, P_no_disease)

# Print the posterior probability
print(f"Posterior Probability of having the disease given a positive test result: {posterior:.4f}")
```

### Explanation of the Code:

1. **Prior (P(Disease))**: Set as `0.01`, representing the initial belief that 1% of people have the disease.
2. **Likelihood (P(Test | Disease))**: Set as `0.95`, which is the probability of a positive test result if the person has the disease.
3. **False Positive Rate (P(Test | No Disease))**: Set as `0.05`, which is the probability of a false positive.
5. **Posterior Calculation**: After running the function, it computes the updated probability of having the disease given a positive test result.

### Interpretation of Results:

- Even with a positive test result and a highly sensitive test (95%), the posterior probability of the person actually having the disease is only about **16.45%**. This is because the prior probability of having the disease is very low (1%), and the false positive rate is still substantial (5%).
  
- This highlights the **importance of the prior probability** in Bayesian reasoning. A seemingly accurate test can still lead to a relatively low probability of disease detection if the prior probability (the initial assumption about how common the disease is) is low.

### Practical Significance of Bayesian Inference:
1. **Medical Diagnosis**: In real-world scenarios, Bayesian inference allows doctors to update their beliefs about a patient's condition based on new evidence (e.g., test results). It helps balance the false positive and false negative rates, leading to more informed decision-making.
   
2. **Improved Predictions**: Instead of relying on static failure rates, Bayesian inference adjusts for each new piece of evidence, making predictions more accurate and reflective of real-world conditions.
   
3. **Reliability Analysis**: In industrial settings, Bayesian methods can be used to estimate the reliability of a system based on prior knowledge and observed failures, continuously improving predictions as more data becomes available.

### Conclusion:
Bayesian probability allows us to update our knowledge and beliefs based on new evidence in a mathematically rigorous way. In the medical diagnosis example, Bayesian inference demonstrates how prior information (e.g., disease prevalence) and new data (e.g., test results) can be combined to make more accurate predictions.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# Function to compute predictions
def predict(X, m, b):
    return m * X + b

# Function to compute Mean Squared Error (MSE)
def compute_loss(X, y, m, b):
    y_pred = predict(X, m, b)
    return np.mean((y_pred - y) ** 2)

# Gradient Descent Function
def gradient_descent(X, y, learning_rate=0.1, max_iter=1000, tolerance=1e-6):
    m, b = -1, 1
    history_m, history_b, losses = [], [], []

    for i in range(max_iter):
        y_pred = predict(X, m, b)
        loss = compute_loss(X, y, m, b)
        losses.append(loss)

        if i > 0 and abs(losses[-2] - loss) < tolerance:
            break

        grad_m = -2 * np.mean((y_pred - y) * X)
        grad_b = -2 * np.mean(y_pred - y)

        m += learning_rate * grad_m
        b += learning_rate * grad_b
        history_m.append(m)
        history_b.append(b)

    return m, b, history_m, history_b, losses

# SciPy Optimization Function
def optimize_with_scipy(X, y):
    def cost_function(params):
        m, b = params
        return compute_loss(X, y, m, b)

    result = minimize(cost_function, [-1, 1])
    m_scipy, b_scipy = result.x
    return m_scipy, b_scipy, result.fun

# Function to plot the results
def plot_convergence(history_m, history_b, m_scipy, b_scipy, losses, scipy_loss):
    plt.figure(figsize=(12, 6))

    # Plot for m and b convergence
    plt.subplot(1, 2, 1)
    plt.plot(history_m, label='m (Gradient Descent)')
    plt.plot(history_b, label='b (Gradient Descent)')
    plt.axhline(y=m_scipy, color='r', linestyle='--', label='m (SciPy)')
    plt.axhline(y=b_scipy, color='b', linestyle='--', label='b (SciPy)')
    plt.xlabel('Iterations')
    plt.ylabel('Value')
    plt.legend()
    plt.title('Parameter Convergence')

    # Plotting loss reduction
    plt.subplot(1, 2, 2)
    plt.plot(losses, label='Gradient Descent Loss')
    plt.axhline(y=scipy_loss, color='b', linestyle='--', label='loss (SciPy)')
    plt.xlabel('Iterations')
    plt.ylabel('MSE Loss')
    plt.legend()
    plt.title('Loss Reduction')

    plt.tight_layout()
    plt.show()

# Sample data
X = np.array([1, 3])
y = np.array([3, 6])

# Hyperparameters
learning_rate = 0.1
max_iter = 1000
tolerance = 1e-6

# Gradient Descent
m, b, history_m, history_b, losses = gradient_descent(X, y, learning_rate, max_iter, tolerance)

# SciPy Optimization
m_scipy, b_scipy, scipy_loss = optimize_with_scipy(X, y)

# Print results
print(f"Gradient Descent - m: {m}, b: {b}")
print(f"SciPy Optimization - m: {m_scipy}, b: {b_scipy}")

# Plot results
plot_convergence(history_m, history_b, m_scipy, b_scipy, losses, scipy_loss)
```

In this solution, we're implementing two optimization methods for finding the best parameters of a simple linear regression model: **Gradient Descent** and **SciPy's `minimize`**. Below is an explanation of the key aspects of the code and the insights gained.

### Code Explanation:

1. **Sample Data**:
   - `X = np.array([1, 3])`: The independent variable values (input data).
   - `y = np.array([3, 6])`: The dependent variable values (target data).

2. **Hyperparameters**:
   - `learning_rate = 0.1`: This controls how large the steps are when updating the parameters during gradient descent.
   - `max_iter = 1000`: Maximum number of iterations for gradient descent.
   - `tolerance = 1e-6`: The stopping criterion for gradient descent. The algorithm stops if the change in loss between two iterations is smaller than this threshold.

3. **Initialize Parameters**:
   - `m, b = -1, 1`: Initial values for the parameters `m` (slope) and `b` (intercept) of the line.

4. **Gradient Descent Optimization**:
   - In each iteration, we:
     - Predict the values of `y` using the current values of `m` and `b`: `y_pred = m * X + b`.
     - Compute the Mean Squared Error (MSE) loss: `loss = np.mean((y_pred - y) ** 2)`.
     - Calculate the gradients with respect to `m` and `b`:
       - `grad_m = -2 * np.mean((y_pred - y) * X)` (gradient with respect to `m`).
       - `grad_b = -2 * np.mean(y_pred - y)` (gradient with respect to `b`).
     - Update the parameters `m` and `b` using the gradients and the learning rate:
       - `m += learning_rate * grad_m`
       - `b += learning_rate * grad_b`
     - Track the parameter values (`history_m` and `history_b`) and loss (`losses`) over iterations.

   - The stopping condition for gradient descent is when the change in the loss between consecutive iterations is less than the `tolerance`.

5. **SciPy Optimization**:
   - Define a cost function `cost_function(params)` that calculates the MSE loss.
   - Use `scipy.optimize.minimize()` to minimize the cost function and find the optimal parameters `m_scipy` and `b_scipy`.

6. **Plotting**:
   - Plot the evolution of the parameters (`m` and `b`) over iterations.
   - Plot the loss reduction over iterations, showing how the optimization process converges.

### Key Insights:

- **Gradient Descent**:
  - Gradient Descent is an iterative optimization method. In this implementation, the model parameters (`m` and `b`) gradually approach their optimal values. The rate of convergence depends on the learning rate and the characteristics of the data.
  - In the plot showing the parameter convergence, we can see how the values of `m` and `b` stabilize over time as the algorithm finds the optimal solution.
  - The plot showing loss reduction illustrates how the MSE loss decreases over time, showing the gradual improvement of the model.

- **SciPy Optimization**:
  - The `minimize()` function from SciPy is a more direct approach to solving the problem. It efficiently finds the optimal values of `m` and `b` without requiring a learning rate or iteration control.
  - The results from SciPy are compared to the final values from Gradient Descent. In most cases, both methods should converge to similar values if the model is well-specified.

- **Comparison**:
  - While Gradient Descent gives insight into how the optimization progresses over time, SciPy's method provides the final optimal values in fewer steps. This shows the efficiency of built-in optimization routines versus manual gradient-based methods.
  - The convergence of both methods should align, showing that both approaches are valid for finding the best-fit line for simple linear regression.

### Final Results and Visuals:

- **Parameter Convergence Plot**:
  - Shows the change in `m` and `b` as the iterations progress. The dashed lines represent the optimal parameters found using SciPy's `minimize` function. This visualization helps understand how gradient descent reaches these optimal values.

- **Loss Reduction Plot**:
  - Tracks how the loss (MSE) decreases as the model's parameters are updated. The loss function converges to a minimum, indicating that the model is fitting the data better with each iteration.

By implementing both Gradient Descent and SciPy's optimization method, we've explored two distinct approaches to solving the same problem. This comparison helps to understand the trade-offs between manual optimization and automated optimization methods.
